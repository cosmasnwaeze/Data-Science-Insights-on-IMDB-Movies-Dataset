## EDA on IMDB Movies Dataset: Using Python to Clean, Explore and Derive insights with the right visualizations from the dataset 

## Data
PosterLink: Link of the poster that imdb using

SeriesTitle: Name of the movie

ReleasedYear: Year at which that movie released

Certificate: Certificate earned by that movie

Runtime: Total runtime of the movie

Genre: Genre of the movie

IMDB Rating: Rating of the movie at IMDB site

Overview: mini story/ summary

Meta_score: Score earned by the movie

Director: Name of the Director

Star1,Star2,Star3,Star4: Name of the Stars

No of votes: Total number of votes

Gross: Money earned by that movie

```python
# importing libraries
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
imdb_df = pd.read_csv(r"C:\Users\LENOVO\Documents\AIML Classwork 3\imdb_top_1000.csv")
```

```python
# Displays the first five rows of data
imdb_df.head()
```

```python
imdb_df.info()
```

```python
shape = imdb_df.shape
shape
```

```python
missing_values = imdb_df.isnull().sum()
missing_values
```

```python
# percentage of misssing values
total_cells = np.prod(shape)
total_missing = missing_values.sum()
percent_missing = (total_missing/total_cells)*100
percent_missing
```

```python
# Drop missing values
imdb_df.dropna(inplace=True)
```

```python
# Verify missing values
print(imdb_df.isnull().sum())
```

```python
imdb_df.drop_duplicates(inplace=True)
```

```python
imdb_df.shape
```

```python
#Rename Series_Title to Movies_Title
imdb_df.rename(columns={"Series_Title": "Movies_Title"}, inplace=True)
imdb_df.head()
```

```python
# Change Gross values to int instead of object
print(imdb_df["Gross"].head(2))
```

```python
imdb_df["Gross"] = imdb_df["Gross"].str.replace(",", "")
print(imdb_df["Gross"].head(2))
```

```python
imdb_df["Gross"] = imdb_df["Gross"].astype('float64')
imdb_df["Gross"] = imdb_df["Gross"].replace(np.nan, 0)
```

```python
imdb_df["Gross"] = imdb_df["Gross"].astype(int)
imdb_df["Gross"].dtype
print(imdb_df["Gross"].head(2))
```

```python
print(imdb_df['Released_Year'].value_counts())
```

```python
imdb_df['Released_Year'] = pd.to_numeric(imdb_df['Released_Year'], errors='coerce')
imdb_df = imdb_df.dropna(subset=['Released_Year'])
```

```python
imdb_df['Released_Year'] = imdb_df['Released_Year'].astype(int)
imdb_df["Released_Year"].dtype
print(imdb_df["Released_Year"].head(10))
```

```python
imdb_df.info()
```

```python
#Top 10 Movies by IMDB Rating
top_10_imdb = imdb_df.nlargest(10, "IMDB_Rating")[["Movies_Title", "IMDB_Rating"]]
print(top_10_imdb)
```

```python
# Top 10 Movies by Gross

top_10_gross = imdb_df.nlargest(10, "Gross")[["Movies_Title", "Gross"]]
print(top_10_gross)
```

```python
# Distribution of IMDB Ratings
plt.figure(figsize=(10, 6))
sns.histplot(imdb_df["IMDB_Rating"], bins=20)
plt.title("Distribution of IMDB Ratings")
plt.show()
```

```python
# Distribution of Gross
plt.figure(figsize=(10, 6))
sns.histplot(imdb_df["Gross"], bins=20)
plt.title("Distribution of Gross")
plt.show()
```

```python
# Correlation between IMDB Rating and Gross
ig_correlation = imdb_df["IMDB_Rating"].corr(imdb_df["Gross"])
print(f"Correlation between IMDB rating and gross: {ig_correlation:.2f}")
```

```python
# Genre with the Highest Average IMDB Rating
genre_imdb_avg = imdb_df.groupby("Genre")["IMDB_Rating"].mean().sort_values(ascending=False)
print(genre_imdb_avg.head())
```

```python
imdb_df.describe()
```

```python
imdb_df.describe().corr()
```

```python
# Calculate the correlation matrix
corr_matrix = imdb_df.describe().corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix of IMDB Dataset')
plt.show()
```

```python
numerical_attributes = ["No_of_Votes", "Meta_score", "IMDB_Rating", "Gross"]
imdb_df[numerical_attributes].hist(figsize = (15, 6), color = "blue", edgecolor = "yellow", layout = (2, 2));
```

```python
# Director with the Most Movies in the Top 10 IMDB Ratings

# Get the top 10 movies by IMDB rating
top_10_movies = imdb_df.nlargest(10, 'IMDB_Rating')

# Count of number of movies for each director in the top 10
director_count = top_10_movies['Director'].value_counts()

# Director with the most movies in the top 10
print(director_count.idxmax(), director_count.max())
```

```python
# Top 10 directors with the most movies in the top 10
top_10_directors = top_10_movies["Director"].value_counts().nlargest(10)

# Print the top 3 directors
print(top_10_directors)
```

```python
# Top 10 Directors with the Most Movies in the Top 10 IMDB Ratings

plt.figure(figsize=(10, 6))
top_10_directors.plot(kind="bar")
plt.title("Top 10 Directors with the Most Movies in the Top 10 IMDB Ratings")
plt.xlabel("Director")
plt.ylabel("Number of Movies")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Analysis of the gross of a movie vs directors.
director_gross_avg = imdb_df.groupby("Director")["Gross"].mean().sort_values(ascending=False)

# Plot the top 10 directors by average gross
plt.figure(figsize=(10, 6))
sns.barplot(x=director_gross_avg.index[:10], y=director_gross_avg.values[:10])
plt.title("Top 10 Directors by Average Gross")
plt.xlabel("Director")
plt.ylabel("Average Gross")
plt.xticks(rotation = 90)
plt.show()
```

```python
# Top voted movies

top_voted = imdb_df.sort_values(["No_of_Votes"], ascending = False)

fig,axs=plt.subplots(figsize=(15,5))
g=sns.barplot(x=top_voted["Movies_Title"][:7],y=top_voted["No_of_Votes"][:7])

g.set_title("Top Voted Movies", weight = "bold")
plt.show()
```

```python
# Top voted movies ratings

fig,axs=plt.subplots(figsize=(15,5))
g=sns.barplot(x=top_voted["Movies_Title"][:7],y=top_voted["IMDB_Rating"][:7])

g.set_title("IMDB Rating of top voted movies", weight = "bold")
plt.show()
```

```python
# No_of_Votes vs Directors

director_votes_avg = imdb_df.groupby("Director")["No_of_Votes"].mean().sort_values(ascending=False)

# Plot the top 10 directors by average No_of_Votes
plt.figure(figsize=(10, 6))
sns.barplot(x=director_votes_avg.index[:10], y=director_votes_avg.values[:10])
plt.title("Top 10 Directors by Average No_of_Votes")
plt.xlabel("Director")
plt.ylabel("Average No_of_Votes")
plt.xticks(rotation = 90)
plt.show()
```

```python
# Combination of Actors with Good IMDB_Rating
star_combination_imdb_avg = imdb_df.groupby(["Star1", "Star2", "Star3", "Star4"])["IMDB_Rating"].mean().reset_index()

# Plot the top 10 combinations by average IMDB_Rating
plt.figure(figsize=(10, 6))
sns.barplot(x="IMDB_Rating", y="Star1", data=star_combination_imdb_avg.nlargest(10, "IMDB_Rating"))
plt.title("Combination of Actors with Good IMDB_Rating")
plt.xlabel("IMDB_Rating")
plt.ylabel("Star Combination")
plt.show()
```

```python
# Combination of Actors with Good Gross
star_combination_gross_avg = imdb_df.groupby(["Star1", "Star2", "Star3", "Star4"])["Gross"].mean().reset_index()

# Plot the top 10 combinations by average Gross
plt.figure(figsize=(10, 6))
sns.barplot(x="Gross", y="Star1", data=star_combination_gross_avg.nlargest(10, "Gross"))
plt.title("Combination of Actors with Good Gross")
plt.xlabel("Gross")
plt.ylabel("Star Combination")
plt.show()
```

```python
# Gross of top rated movies
fig, axs = plt.subplots(figsize=(15, 5))
g = sns.barplot(x=top_voted['Movies_Title'][:7], y=top_voted['Gross'][:7])
g.set_title("Gross by Top Rated Movies", weight="bold")
plt.show()
```

```python
# Top movies by Gross
highest_earning = imdb_df.sort_values(["Gross"], ascending = False)
fig, axs = plt.subplots(figsize=(23, 5))
g = sns.barplot(x=highest_earning["Movies_Title"][:7], y=highest_earning["Gross"][:7])
g.set_title("Movies with Highest Gross", weight="bold")
plt.show()
```

```python
# Maximum movies released through the years-
fig, axs = plt.subplots(figsize=(20, 5))
g = sns.barplot(x=imdb_df["Released_Year"].value_counts()[:10].index, y=imdb_df["Released_Year"].value_counts()[:10])
g.set_title("Maximum Series Released through the years", weight="bold")
g.set_xlabel("Years")
plt.show()
```

```python
# Mostly occurred Directors
fig,axs=plt.subplots(figsize=(20,5))
g=sns.lineplot(x=imdb_df["Director"].value_counts()[:10].index,y=imdb_df["Director"].value_counts()[:10])
g.set_title("Mostly Occurred Directors", weight = "bold")
g.set_xlabel("Directors", weight = "bold")
plt.xticks(rotation=90)
plt.show()

```
